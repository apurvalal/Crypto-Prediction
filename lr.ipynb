```python
import pandas
```

```python
BTC= pandas.read_csv("/archive/bitcoin_dataset.csv")
library(caret)

BTC= na.omit(BTC)
library(corrplot)
cor = cor(BTC[2:12])
corrplot(cor, method = "pie")
#btc_market_cap, btc_hash_rate, btc_difficulty are most correlated with btc_market_price
```

```python
plot(BTC$Date, BTC$btc_market_price)
```

```python
mod<-lm(btc_market_price~btc_market_cap, data=BTC)
mod1<-lm(btc_market_price~btc_estimated_transaction_volume_usd, data=BTC)
summary(mod)
#intercept is 3.23, which means that when the total USD value of bitcoin supply in circulation is 0, the average USD market price across major bitcoin exchanges is predicted to be 3.23$.
# slope is  6.029,  it tells us that we predict the average USD market price across major bitcoin exchanges to increase by 6$ for every additional dollar increase in the total USD value of bitcoin supply in circulation 
summary(mod1)
#intercept is 6.4, which means that when the estimated transaction value in USD value is 0, the average USD market price across major bitcoin exchanges is predicted to be 6.4$.
# slope is  3.6,  it tells us that we predict the average USD market price across major bitcoin exchanges to increase by 3.6$ for every additional dollar increase in the estimated transaction value.
```

```python
plot(BTC$btc_market_price,BTC$btc_market_cap)
plot(BTC$btc_market_price,BTC$btc_estimated_transaction_volume_usd)
```

```python
set.seed(1)
train.index<-sample(1:nrow(BTC),0.70*nrow(BTC), replace=FALSE)
train <- BTC[train.index, ]
test  <- BTC[-train.index,]
```

```python
#Training the model on train dataset and  predicting on test.
model1 <- lm(btc_market_price~btc_trade_volume , train)
summary(model1)
p1 <- predict(model1,test)
head(p1)
error1 <- p1 - test[["btc_market_price"]]
sqrt(mean(error1^2))
plot(p1)
#RMSE=510.7136
```

```python
model2 <- lm(btc_market_price~btc_market_cap, train)
summary(model2)
p2 <- predict(model2,test)
head(p2)
error2 <- p2 - test[["btc_market_price"]]
sqrt(mean(error2^2))
plot(p2)
#RMSE=47.95438
#best model for predicting the market price of bitcoin
```

```python
# PCA on Bitcoin market price 
library(FactoMineR)
pc1<-PCA(BTC[,2:24],scale.unit = TRUE, ncp = 23, graph = TRUE)
summary(pc1)
a<-dimdesc(pc1,axes = c(1:2))
a$Dim.1
a$Dim.2
```

```python
pc<-prcomp(BTC[,c("btc_market_price","btc_miners_revenue")], center = T, scale=T)
summary(pc)
head(pc$x)
#we can see that the first principal component explains about 61.6% of the total variation, and the second principal component an additional 10.98%. 
#btc_n_transactions_total and btc_blocks_size  are the most contributing variables
```

```python
model3<-lm(btc_market_price~btc_n_transactions_total, data=BTC)
summary(model3)
p3 <- predict(model3,test)
error3 <- p3 - test[["btc_market_price"]]
sqrt(mean(error3^2))
#RMSE=616.3098
#intercept is -1.16, which means that when total number of transactions is 0, the average USD market price across major bitcoin exchanges is predicted to be -1.
# slope is  9.8,  it tells us that we predict the average USD market price across major bitcoin exchanges to increase by 9.8$ for every additional dollar increase in total number of transactions.
```

```python
model4<-lm(btc_market_price~btc_n_transactions_total+btc_avg_block_size+btc_difficulty+btc_output_volume, data=BTC)
summary(model4)
#Multiple R-squared:  0.9112,	Adjusted R-squared:  0.9111 , all independent variables affect the dependent one.
p4 <- predict(model4,test)
error4 <- p4 - test[["btc_market_price"]]
sqrt(mean(error4^2))
#RMSE=295.8589
```
